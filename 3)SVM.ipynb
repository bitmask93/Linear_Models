## Support Vector Machines


- Support-vector machines(SVMs) are supervised learning models capable of performing both Classification as well as Regression analysis.


- Given a set of training examples each belonging to one or the other two categories, an SVM training algorithm builds a model that assigns new examples to one category or the other.


- SVM is a non-probabilistic binary linear classification algorithm ie given a training instance, it will not output a probability distribution over a set of classes rather it will output the most likely class that the observation should belong to.


- However methods such as Platt scaling exist to use SVM in a probabilistic classification setting. Platt scaling or Platt Calibration  is a way of transforming the outputs of a classification model into a probability distribution over classes. Platt scaling works by fitting a logistic regression model to a classifier's scores.


- In addition to performing linear classification, SVMs can efficiently perform a non-linear classification using what is called the kernel trick, implicitly mapping the inputs into high-dimensional feature spaces.


- In the case of support-vector machines, a data point is viewed as a $p-dimensional$ vector (a list of $p$ numbers), and we want to know whether we can separate such points with a $(p-1)$-dimensional hyperplane. 


- There are many hyperplanes that might classify the data. One reasonable choice as the best hyperplane is the one that represents the largest separation, or margin, between the two classes.


-  So we choose the hyperplane so that the distance from it to the nearest data point on each side is maximized. Such a hyperplane is called as a *maximum-margin hyperplane* and the linear classifier it defines is called as a margin maximizing classifier.


<img src="images/margin-hyperplane.png" width="600" height="600">


- Here we can see three hyperplanes $H_{1}$, $H_{2}$, $H_{3}$. Here $H_{1}$ does not separate the classes, but $H_{2}$ and $H_{3}$ does but the margin is maximum for $H_{3}$.


-  A good separation is achieved by the hyperplane that has the largest distance to the nearest training-data point of any class.


- The larger the margin, the lower the generalization error of the classifier.


- Consider we are given the training examples : $(x_{1},y_{1}),....,(x_{n},y_{n})$

- Here $y_{i}$ belongs to either +1 or -1 indicating the class to which $y_{i}$ belongs to. and each $x_{i}$ is a *p-dimensional* real vector.


- The objective is to find a hyperplane that maximises the margin that divides the group of points $x_{i}$ for which $y_{i} = 1$ and for which $y_{i} = -1$.


- Any hyperplane can be written as the set of points $x$ satisfying

$$ w.x-b = 0$$

- Where $w$ is the normal vector to the hyperplane. Here $w$ need not be normalized. The parameter $\frac{b}{\left \| w \right \|}$ determines the offset of the hyperplane from the origin along the normal vector $w$


#### Hard Margin SVM

- If the data is linearly separable, we can select two parallel hyperplanes that separates the two classes of data, so that the distance between them is as large as possible.


- The region bounded by these two hyperplanes is called the **margin**, and the maximum-margin hyperplane is the hyperplane that lies halfway between them.


- These hyperplanes can be describes with following equations : 

$$w.x - b = 1$$
- Anything on or above this boundary is of class with label 1.


$$w.x - b = -1$$
- Anything on or below this boundary is of the class with label âˆ’1


- Geometrically the distance between these two hyperplanes is given as $\frac{2}{\left \| w \right \|}$, so to maximize the distance between the planes, we want to minimize $\left \| w \right \|$ .


- The distance is computed using the distance from a point to a plane equation. To prevent data points from falling into the margin, we add the following constraint: for each $i$ either:

$$w.x - b \geq 1$$$$y_{i} = 1$$

$$OR$$


$$w.x - b \leq -1$$$$y_{i} = -1$$


<img src="images/svm.png" width="500" height="500">


- These constraints state that each data point must lie on the correct side of the margin.


- The above constraints can be rewritten as 

$$y_{i}(w.x_{i}-b)\geq1$$ $$\forall 1\leq i\leq n$$


- The optimization problem can be written as :


- Minimize $\left \| w \right \|$ subject to $y_{i}(w.x_{i} - b)\geq 1$ for $i = 1,2,3....,n$


- The $w$ and $b$ that solve this problem determine the classifier : given $x$ $->$ $sign(w.x - b)$.


- Maximum-margin hyperplane is completely determined by those $x_{i}$ which is nearest to it. These $x_{i}$ are called *Support vectors*. ie they are the data points on the margin.


#### Soft-margin SVM

- Hard-margin SVM requires data to be linearly separable. But in real-world this does not happen always. So we use the hinge-loss function which is given as :


$$max(0 , 1-y_{i}(w.x - b)$$


- $y_{i}$ is the $i^{th}$ target(1 or -1) and $(w.x - b)$ is the $i^{th}$ output


- This function is $0$, if $x_{i}$ lies on the correct side of the margin. For data on the wrong side of the margin, the function's value is proportional to the distance from the margin.


- We need to minimize the following : 

$$ \begin{bmatrix}
\frac{1}{n}\sum_{i=1}^{n}max\left ( 0 , 1 - y_{i}\left ( w.x_{i} - b \right ) \right )
\end{bmatrix} + \lambda\left \| w \right \|^{2}$$


- $\lambda$ determines the tradeoff between increasing the margin size and ensuring that the $x_{i}$ lie on the correct side of the margin.


- For sufficiently small values of $\lambda$ , the second term in the loss function will become negligible, hence, it will behave similar to the hard-margin SVM.

## Computing the SVM classifier

- In SVM, we try to minimize the following expression :

$$ \begin{bmatrix}
\frac{1}{n}\sum_{i=1}^{n}max\left ( 0 , 1 - y_{i}\left ( w.x_{i} - b \right ) \right )
\end{bmatrix} + \lambda\left \| w \right \|^{2}$$


**Primal**


- Minimizing the above equation can be rewritten as a constrained optimization problem with a differentiable objective function in the following way:


- For each $i \in \left \{1,....,n  \right \}$, we introduce a variable $\zeta_{i} = max(0,1-y_{i}(w.x_{i}-b))$.


- $\zeta_{i}$ is the smallest nonnegative number satisfying $y_{i}(w.x_{i} - b)\geq 1 - \zeta_{i}$


- Now we can rewrite the optimization problem as :

$${\displaystyle {\text{minimize }}{\frac {1}{n}}\sum _{i=1}^{n}\zeta _{i}+\lambda \|w\|^{2}}$$

$${\displaystyle {\text{subject to }}y_{i}(w\cdot x_{i}-b)\geq 1-\zeta _{i}\,{\text{ and }}\,\zeta _{i}\geq 0,\,{\text{for all }}i.}$$
 
 
 **Dual**
 
 - By solving for the Lagrangian dual of the above problem, we can get the simplified problem.


$${\displaystyle {\text{maximize}}\,\,f(\alpha_{1}\ldots \alpha_{n})=\sum _{i=1}^{n}\alpha_{i}-{\frac {1}{2}}\sum _{i=1}^{n}\sum _{j=1}^{n}y_{i}\alpha_{i}(x_{i}\cdot x_{j})y_{j}\alpha_{j}}$$

$${\displaystyle{\text{subject to }}\sum _{i=1}^{n}\alpha_{i}y_{i}=0,\,{\text{and }}0\leq \alpha_{i}\leq {\frac {1}{2n\lambda }}\;{\text{for all }}i}$$


- $\alpha_{i}$ is defined such that : 

$${\displaystyle {{w}}=\sum _{i=1}^{n}\alpha_{i}y_{i}{{x}}_{i}}$$


-  ${\displaystyle \alpha_{i}=0}$ exactly when ${\displaystyle {{x}}_{i}}$ lies on the correct side of the margin.


- ${\displaystyle 0<\alpha_{i}<(2n\lambda )^{-1}}$ when ${\displaystyle { {x}}_{i}}$ lies on the margin's boundary.


- It follows that ${\displaystyle {w}}$ can be written as a linear combination of the support vectors.


- The offset, ${\displaystyle b} $, can be recovered by finding an ${\displaystyle {{x}}_{i}}$ on the margin's boundary and solving the following : 

$${\displaystyle y_{i}({{w}}\cdot {{x}}_{i}-b)=1\iff b={{w}}\cdot {{x}}_{i}-y_{i}}$$


- ${\displaystyle y_{i}^{-1}=y_{i}}$ since ${\displaystyle y_{i}=\pm 1}$

**Kernel trick**


- Suppose now that we would like to learn a nonlinear classification rule which corresponds to a linear classification rule for the transformed data points ${\displaystyle \varphi ({ {x}}_{i})}$


- Say we have a kernel function $k$ which satisfies ${\displaystyle k({{x}}_{i},{{x}}_{j})=\varphi ({{x}}_{i})\cdot \varphi ({{x}}_{j})}$


- Some common kernels include :

    - > Polynomial (homogeneous) : ${\displaystyle k({{x_{i}}},{{x_{j}}})=({{x_{i}}}\cdot {{x_{j}}})^{d}}$
    
    - > Polynomial (inhomogeneous) : ${\displaystyle k({{x_{i}}},{{x_{j}}})=({{x_{i}}}\cdot {{x_{j}}}+1)^{d}}$
    
    - > Gaussian radial basis function : ${\displaystyle k({{x_{i}}},{{x_{j}}})=\exp(-\gamma \|{{x_{i}}}-{{x_{j}}}\|^{2})}$ for ${\displaystyle \gamma >0}$ Sometimes parametrized using ${\displaystyle \gamma =1/(2\sigma ^{2})}$
    
    - > Hyperbolic tangent : ${\displaystyle k({{x_{i}}},{{x_{j}}})=\tanh(\kappa {{x_{i}}}\cdot {{x_{j}}}+c)}$ for some (not every) ${\displaystyle \kappa >0}$ and ${\displaystyle c<0}$
    

-  The classification vector ${\displaystyle {{w}}}$ in the transformed space satisfies :


$${\displaystyle {{w}}=\sum _{i=1}^{n}\alpha_{i}y_{i}\varphi ({{x}}_{i})}$$


- Where, the ${\displaystyle \alpha_{i}}$ are obtained by solving the optimization problem.


$${\displaystyle {\begin{aligned}{\text{maximize}}\,\,f(\alpha_{1}\ldots \alpha_{n})&=\sum _{i=1}^{n}\alpha_{i}-{\frac {1}{2}}\sum _{i=1}^{n}\sum _{j=1}^{n}y_{i}\alpha_{i}(\varphi ({{x}}_{i})\cdot \varphi ({{x}}_{j}))y_{j}\alpha_{j}\\&=\sum _{i=1}^{n}\alpha_{i}-{\frac {1}{2}}\sum _{i=1}^{n}\sum _{j=1}^{n}y_{i}\alpha_{i}k({{x}}_{i},{{x}}_{j})y_{j}\alpha_{j}\\\end{aligned}}}$$



$${\displaystyle {\text{subject to }}\sum _{i=1}^{n}\alpha_{i}y_{i}=0,\,{\text{and }}0\leq \alpha_{i}\leq {\frac {1}{2n\lambda }}\;{\text{for all }}i}$$


- The coefficients ${\displaystyle \alpha_{i}}$ can be solved for using quadratic programming.


- We can find some index ${\displaystyle i}$ such that ${\displaystyle 0<\alpha_{i}<(2n\lambda )^{-1}}$, so that ${\displaystyle \varphi ({{x}}_{i})}$ lies on the boundary of the margin in the transformed space, and then solve:


$${\displaystyle {\begin{aligned}b={{w}}\cdot \varphi ({{x}}_{i})-y_{i}&=\left[\sum _{j=1}^{n}\alpha_{j}y_{j}\varphi ({{x}}_{j})\cdot \varphi ({{x}}_{i})\right]-y_{i}\\&=\left[\sum _{j=1}^{n}\alpha_{j}y_{j}k({{x}}_{j},{{x}}_{i})\right]-y_{i}\end{aligned}}}$$

- And :

$$
{\displaystyle {{z}}\mapsto \operatorname {sgn}({{w}}\cdot \varphi ({{z}})-b)=\operatorname {sgn} \left(\left[\sum _{i=1}^{n}\alpha_{i}y_{i}k({{x}}_{i},{{z}})\right]-b\right).}$$

## Sub-Gradient Descent for SVM

- Why can't we use 0-1 Loss in SVM instead of Hinge Loss?


- 0-1 loss function is flat so it doesn't converge well. Also it is not differentiable at 0. SVM is based on solving an optimization problem that maximize the margin between classes. So in this context a convex loss function is preferable so we can use several general convex optimization methods. The 0-1 loss function is not convex so it is not very useful either. 


- SVM objective function is nothing but Hinge loss with $l_{2}$ regularization.


- SVM Loss Function : 

$$$$


- This function is not differentiable at $x = 1$


- Derivative of hinge loss $L(m) = max(0,1-m)$ :

$${L}'(m) = \left\{\begin{matrix}
0 & m>1\\ 
-1 & m<1\\ 
undefined &m=1 
\end{matrix}\right.$$


- We need gradient with respect to parameter vector $w âˆˆ R^{d}$. For simplicity, we will not consider the bias term $b$.

$$\bigtriangledown _{w} L(y_{i}w^{T}x_{i}) = {L}'(y_{i}w^{T}x_{i})y_{i}x_{i}$$

$$= \left ( \left\{\begin{matrix}
0 & y_{i}w^{T}x_{i}>1\\ 
-1 & y_{i}w^{T}x_{i}<1\\ 
undefined &y_{i}w^{T}x_{i}=1 
\end{matrix}\right. \right )$$

$$ \bigtriangledown _{w} L(y_{i}w^{T}x_{i}) =  \left\{\begin{matrix}
0 & y_{i}w^{T}x_{i}>1\\ 
-y_{i}x_{i} & y_{i}w^{T}x_{i}<1\\ 
undefined &y_{i}w^{T}x_{i}=1 
\end{matrix}\right. $$

$$\bigtriangledown _{w}L(w) = \bigtriangledown _{w}\left ( \frac{1}{n}\sum_{i=1}^{n}L(y_{i}w^{T}x_{i}) + \lambda\left \| w \right \|^{2} \right)$$

$$= \frac{1}{n}\sum_{i = 1}^{n}\bigtriangledown _{w}L(y_{i}w^{T}x_{i}) + 2\lambda w$$

$$= \left\{\begin{matrix}
\frac{1}{n}\sum_{i:y_{i}w^{T}x_{i}<1}(-y_{i}x_{i}) + 2\lambda w &y_{i}w^{T}x_{i} \neq 1 \\ 
undefined & otherwise 
\end{matrix}\right.$$


- So the Gradient of SVM Objective function is :

$$\bigtriangledown _{w}L(w) = \frac{1}{n}\sum_{i:y_{i}w^{T}x_{i}<1}(-y_{i}x_{i}) + 2 \lambda w$$

- When $y_{i}w^{T}x_{i} \neq 1$ for all $i$, **otherwise it is undefined**


- Why we shouldnâ€™t care about the points of nondifferentiability:

    - > If we start with a random $w$, the chances that we get $y_{i}w^{T}x_{i} = 1$ is very rare.
    
    - > Even if we did reach such a point, we could add a small $\epsilon$ to miss the point where our algorithm becomes 1.

**Subgradient Of SVM Loss Function**:

$$\bigtriangledown _{w}L(w) = \left\{\begin{matrix}
-y_{i}x_{i} & y_{i}w^{T}x_{i} <1 \\ 
0 & y_{i}w^{T}x_{i} \geq 1
\end{matrix}\right.$$

$$= -y_{i}x_{i} I_{\left \{ y_{i}w^{T}x_{i}<1 \right \}}$$


- Where $I_{\left \{ y_{i}w^{T}x_{i}<1 \right \}}$ is called the **Indicator Function**

$$I_{\left \{ y_{i}w^{T}x_{i}<1 \right \}} = \left\{\begin{matrix}
1 & y_{i}w^{T}x_{i}<1\\ 
0 & y_{i}w^{T}x_{i}\geq 1 
\end{matrix}\right.$$

- The subgradient of Cost Function can be written as :

$$\bigtriangledown _{w}L(w) = \frac{1}{n}\sum_{i=1}^{n}(-y_{i}x_{i}I_{\left \{ y_{i}w^{T}x_{i}<1 \right \}}) + 2 \lambda w$$

### Platt Scaling :

Refer : http://citeseerx.ist.psu.edu/viewdoc/download;jsessionid=4DDE09F156D9C7B25755CE1B2B593F2F?doi=10.1.1.41.1639&rep=rep1&type=pdf

- In machine learning, Platt scaling or Platt calibration is a way of transforming the outputs of a classification model into a probability distribution over classes.


- Platt scaling works by fitting a logistic regression model to a classifier's scores.


- Consider that we have a Binary classification problem : So for a given input $x$, we want to determine if they belong to either of two classes $+1$ or $-1$.


- Let's say the classification problem will be solved by a real-valued function $f$, by predicting a class label $y = sign(f(x))$.



- Platt Scaling produces probability estimates by applying a logistic transformation of the classifier scores $f(x)$ :

$${\mathrm  {P}}(y=1|x)={\frac  {1}{1+\exp(Af(x)+B)}}$$

- The parameters A and B are estimated using maximum likelihood estimation from a training set $(f_{i} , y_{i})$


-  First, let us define a new training set $(f_{i},t_{i})$ where $t_{i}$ are target probabilities defined as :


$$t_{i} = \frac{y_{i}+1}{2}$$

- The parameters $A$ and $B$ are found by minimizing the negative log likelihood of the training data, which is a cross-entropy error function : 

$$min -\sum_{i}^{}t_{i}log(p_{i}) + (1-t_{i})log(1-p_{i})$$

- Where : 
$$p_{i} = \frac{1}{1+\exp(Af_{i}+B)}$$


- The minimization is a two-parameter minimization. Hence, it an be performed using any number of optimization algorithms.


- The probability of corret label an be derived using Bayes' rule. Let us choose a uniform uninformative prior over probabilities of corret label. Now, let us observe $N_{+}$ positive examples. The MAP estimate for the target probability of positive examples$(y = +1)$ is:

$$t_{{+}}={\frac  {N_{{+}}+1}{N_{{+}}+2}}$$

- Similarly, if there are $N_{-}$ negative examples, then the MAP estimate for the target probability of negative examples$(y = -1)$ is :

$$t_{{-}}={\frac  {1}{N_{{-}}+2}} $$

## Simpler Implementation of Sub-Gradient Descent

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from IPython import display
from sklearn.metrics import accuracy_score
from numpy.linalg import norm
```

```python
X = np.random.rand(1000,2)
y = 2 * X[:, 0] + -3 * X[:, 1] # True Weights = [3,-3]
y = np.round(1/(1 + np.exp(-y)))

for i in range(len(y)):
    if(y[i]==0):
        y[i] = -1
    else:
        y[i] = 1

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, s=20)
#plt.show()
plt.xticks([])
plt.yticks([])
plt.savefig('data.png')
```

$$L(w) = \frac{1}{n}\sum_{i=1}^{n}max(0,1-y_{i}[w^{T}x_{i} - b]) + \lambda \left \| x \right \|^{2}$$


$$\bigtriangledown _{w}L(w) = \frac{1}{n}\sum_{i=1}^{n}(-y_{i}x_{i}I_{\left \{ y_{i}w^{T}x_{i}<1 \right \}}) + 2 \lambda w$$

```python
def hinge_loss(x, y, w, lambdh):
    b = np.ones(x.shape[0]) #Intercept term: Initialize with ones.
    distances = 1 - y * (np.dot(x, w)-b)
    distances[distances < 0] = 0  # equivalent to max(0, distance)
    loss = np.sum(distances) / x.shape[0]
    
    # calculate cost
    hinge_loss = lambdh * np.dot(w, w) + loss
    return hinge_loss

def get_grads(x,y,w,lambdh):
    b = np.ones(x.shape[0]) #Intercept or Bias : Initialize with one.
    grad_arr = []
    I = y*(np.dot(x, w)-b)  #Indicator Function
    
    for i in range(len(I)):
        if(I[i]<1):
            I[i] = 1
        else:
            I[i] = 0
            
    for i in range(x.shape[0]):
        grad_arr.append(-y[i]*x[i]*I[i])
            
    grads = np.sum(grad_arr, axis=0)/x.shape[0] + 2*lambdh*w
    return grads
```

```python
#mini-batch SGD
np.random.seed(42)
w = np.array([1,-6]) #Initial Weights
eta= 0.1 # learning rate
n_iter = 4000 # Number of Iterations
batch_size = 128
loss = np.zeros(n_iter)
lambdh = 0.0001  #Change the Margin size and see the difference.
es = 0.09 #For early stopping 

for i in range(n_iter):
    ind = np.random.choice(X.shape[0], batch_size)
    loss[i] = hinge_loss(X[ind, :], y[ind], w, lambdh)
    w = w - eta*get_grads(X[ind, :], y[ind], w, lambdh)
    if(i%10==0):
        plt.plot(loss[:i])
        plt.grid()
        ymin, ymax = plt.ylim()
        plt.ylim(0, ymax)
        plt.xlabel('Iterations')
        plt.ylabel('Loss')
        display.clear_output(wait=True)
        plt.show()
    
    if(loss[i]<es):
        break
```

```python
print("mini-batch SGD Converged in : ",i+1)
print("FInal Weights",w)
print('Final Loss = ',hinge_loss(X,y,w,lambdh))

#Get The predictions
y_predicted = np.array([])
for i in range(X.shape[0]):
    yp = np.sign(np.dot(w, X[i])) #model
    y_predicted = np.append(y_predicted, yp)

print("Accuracy = ",accuracy_score(y,y_predicted))
```

- Here if we reduce the margin size, the number of misclassifications becomes small. Because here from the figure we can clearly see that between the two classes the separating distance is small.

#### SGD With Momentum

```python
np.random.seed(42)
w = np.array([1,-6]) #Initial Weights

eta = 0.05 # learning rate
alpha = 0.9 # momentum

h = np.zeros_like(w)  #This is the additional vector

es = 0.09 #For early stopping 


n_iter = 4000
batch_size = 128

lambdh = 0.0001

loss = np.zeros(n_iter)


for i in range(n_iter):
    ind = np.random.choice(X.shape[0], batch_size)
    loss[i] = hinge_loss(X[ind, :], y[ind], w, lambdh)
    
    h = alpha*h + eta*get_grads(X[ind, :], y[ind], w, lambdh)
    
    w = w-h
    
    if(i%10==0):
        plt.plot(loss[:i])
        plt.grid()
        ymin, ymax = plt.ylim()
        plt.ylim(0, ymax)
        plt.xlabel('Iterations')
        plt.ylabel('Loss')
        display.clear_output(wait=True)
        plt.show()
        
        
    if(loss[i]<es):
        break
```

```python
print("mini-batch SGD Converged in : ",i+1)
print("FInal Weights",w)
print('Final Loss = ',hinge_loss(X,y,w,lambdh))

#Get The predictions
y_predicted = np.array([])
for i in range(X.shape[0]):
    yp = np.sign(np.dot(w, X[i])) #model
    y_predicted = np.append(y_predicted, yp)

print("Accuracy = ",accuracy_score(y,y_predicted))
```

#### RMSProp

```python
np.random.seed(42)
w = np.array([1,-6]) #Initial Weights

eta = 0.05 # learning rate
alpha = 0.9 # momentum
gt = None
eps = 1e-8
es = 0.09 #For early stopping 

lambdh = 0.0001

n_iter = 4000
batch_size = 128
loss = np.zeros(n_iter)

for i in range(n_iter):
    ind = np.random.choice(X.shape[0], batch_size)
    loss[i] = hinge_loss(X[ind, :], y[ind], w, lambdh)
    
    gt = get_grads(X[ind, :], y[ind], w, lambdh)**2
    gt = alpha*gt +(1-alpha)*gt
    
    w = w-(eta*get_grads(X[ind, :], y[ind], w, lambdh))/np.sqrt(gt+eps)
    
    if(i%10==0):
        plt.plot(loss[:i])
        plt.grid()
        ymin, ymax = plt.ylim()
        plt.ylim(0, ymax)
        plt.xlabel('Iterations')
        plt.ylabel('Loss')
        display.clear_output(wait=True)
        plt.show()
        
    if(loss[i]<es):
        break
```

```python
print("mini-batch SGD Converged in : ",i+1)
print("FInal Weights",w)
print('Final Loss = ',hinge_loss(X,y,w,lambdh))

#Get The predictions
y_predicted = np.array([])
for i in range(X.shape[0]):
    yp = np.sign(np.dot(w, X[i])) #model
    y_predicted = np.append(y_predicted, yp)

print("Accuracy = ",accuracy_score(y,y_predicted))
```

```python
np.random.seed(42)
w = np.array([1,-6]) #Initial Weights

eta = 0.05 # learning rate
alpha = 0.9 # momentum
gt = None
vt = 0
beta = 0.0005

eps = 1e-8
es = 0.09 #For early stopping 

lambdh = 0.0001

n_iter = 4000
batch_size = 128
loss = np.zeros(n_iter)

for i in range(n_iter):
    ind = np.random.choice(X.shape[0], batch_size)
    loss[i] = hinge_loss(X[ind, :], y[ind], w, lambdh)
    
    gt = get_grads(X[ind, :], y[ind], w, lambdh)**2
    
    vt = (beta*vt) + (1-beta)*gt 
    
    w = w-(eta*get_grads(X[ind, :], y[ind], w, lambdh))/np.sqrt(vt)+eps
    
    if(i%10==0):
        plt.plot(loss[:i])
        plt.grid()
        ymin, ymax = plt.ylim()
        plt.ylim(0, ymax)
        plt.xlabel('Iterations')
        plt.ylabel('Loss')
        display.clear_output(wait=True)
        plt.show()
        
    if(loss[i]<=es):
        break
```

```python
def platt_calib_parms(out,target,prior1,prior0):
    #out = array of SVM outputs
    #target = array of booleans: is ith example a positive example?
    #prior1 = number of positive examples
    #prior0 = number of negative examples
    A = 0
    B = np.log((prior0+1)/(prior1+1))
    hiTarget = (prior1+1)/(prior1+2)

    loTarget = 1/(prior0+2)
    lambdah = 1e-3
    olderr = 1e300

    pp = np.zeros(prior0+prior1)      # temp array to store urrent estimate of probability of examples

    for i in range(len(pp)):
        pp[i] = (prior1+1)/(prior0+prior1+2)
    
    count = 0
    for it in range(1,100):
        a = 0
        b = 0 
        c = 0
        d = 0
        e = 0
        #compute Hessian & gradient of error funtion with respet to A & B
        for i in range(len(target)):
            if (target[i]==1):
                t = hiTarget
            else:
                t = loTarget
            d1 = pp[i]-t
            d2 = pp[i]*(1-pp[i])
            a += out[i]*out[i]*d2
            b += d2
            c += out[i]*d2
            d += out[i]*d1
            e += d1
        #If gradient is really tiny, then stop
        if (abs(d) < 1e-9 and abs(e) < 1e-9):
            break
        oldA = A
        oldB = B
        err = 0
    
        #Loop until goodness of fit inreases
    
        while (1):
            det = (a+lambdah)*(b+lambdah)-c*c
            if (det == 0):  #if determinant of Hessian is zero,inrease stabilizer
                lambdah *= 10
                ontinue
            A = oldA + ((b+lambdah)*d-c*e)/det
            B = oldB + ((a+lambdah)*e-c*d)/det
        
            #Now, ompute the goodness of fit
            err = 0
            for i in range(len(target)):
                p = 1/(1+np.exp(out[i]*A+B))
                pp[i] = p
            
                #At this step, make sure log(0) returns -200
                err -= t*np.log(p)+(1-t)*np.log(1-p)
        
            if (err < olderr*(1+1e-7)):
                lambdah *= 0.1
                break
            #error did not derease: inrease stabilizer by fator of 10 & try again
            lambdah *= 10
            if (lambdah >= 1e6):  # Something is broken. Give up
                break
            diff = err-olderr
            sale = 0.5*(err+olderr+1)
            if (diff > -1e-3*sale and diff < 1e-7*sale):
                count+=1
            else:
                count = 0
            olderr = err
            if(count==3):
                break
    return A,B
```
